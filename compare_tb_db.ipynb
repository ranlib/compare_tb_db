```python
CDC[["Gene","Mutation"]] = CDC.Variants.str.split("_", expand=True)
CDC["Phenotype"] = CDC["Phenotype"].replace("S","-S")
CDC[["Drug","Confers"]] = CDC.Phenotype.str.split("-", expand=True)
```

```python
CDC.head()
```

```python
CDC_mutation_not_in_London = CDC.loc[ ~CDC.Mutation.isin(London.Mutation) ].copy()
```

```python
CDC_mutation_not_in_London.info()
```

```python
CDC_mutation_not_in_London["WHO_Comment"] = CDC_mutation_not_in_London["WHO_Comment"].replace("WHO:","", regex=True)
CDC_mutation_not_in_London_small = CDC_mutation_not_in_London.drop(["Variants", "Phenotype"],axis=1)
CDC_mutation_not_in_London_small.to_csv("CDC_mutation_not_in_London_small.csv",index=False)
```
